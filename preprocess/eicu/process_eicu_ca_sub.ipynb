```python
import numpy as np
import pandas as pd

df=pd.read_csv("eicu_ca_sub.csv")

df_new=pd.DataFrame(columns=['patientunitstayid', 'timeLV', 'vHR','vRR', 'vsbp', 'vdbp','vmbp','vspo2'])

print(df)
```

```python
my_index=0
flag=1
for i in df.index :
    if(flag==df['timeLV'][i]):
            df_new.loc[my_index] = [df['patientunitstayid'][i],
                                    df['timeLV'][i],
                                    df['vHR'][i],
                                    df['vRR'][i],
                                    df['vsbp'][i],
                                    df['vdbp'][i],
                                    df['vmbp'][i],
                                    df['vspo2'][i]
                                   ]       
            my_index=my_index+1 

            print("flag: timeLV: " ,flag, df['timeLV'][i] , df['patientunitstayid'][i])
            flag=flag+1
            if (flag>24):
                   flag=1          
    else :
        while(flag!=df['timeLV'][i]): 
            
                if(flag==24):
                    df_new.loc[my_index] = [df['patientunitstayid'][i-1],
                                            flag,
                                            "",
                                            "",
                                            "",
                                            "",
                                            "",
                                            ""                   
                                           ]           
                    print("flag: timeLV: " ,flag, df['timeLV'][i-1], df['patientunitstayid'][i-1])
                    flag=flag+1
                    if (flag>24):
                           flag=1  
                    my_index=my_index+1
                else:
                    df_new.loc[my_index] = [df['patientunitstayid'][i],
                                            flag,
                                            "",
                                            "",
                                            "",
                                            "",
                                            "",
                                            ""                   
                                           ]           
                    print("flag: timeLV: " ,flag, df['timeLV'][i], df['patientunitstayid'][i])
                    flag=flag+1
                    if (flag>24):
                           flag=1  
                    my_index=my_index+1
            
        df_new.loc[my_index] = [df['patientunitstayid'][i],
            df['timeLV'][i],
            df['vHR'][i],
            df['vRR'][i],
            df['vsbp'][i],
            df['vdbp'][i],
            df['vmbp'][i],
            df['vspo2'][i]
           ]   
        
        print("flag: timeLV: " ,flag, df['timeLV'][i], df['patientunitstayid'][i])
        flag=flag+1
        if (flag>24):
               flag=1  
        my_index=my_index+1
```
