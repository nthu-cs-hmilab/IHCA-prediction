```python
from numpy import expand_dims
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.preprocessing.image import ImageDataGenerator
import os
import cv2
from matplotlib import pyplot
import numpy as np

imgs = []

for filename in os.listdir(r'C:\Users\Tony\my_final_report\cxr_ca_train_multi_selected_1'):
        #print(filename) #just for test
        #img is used to store the image data 
        img = cv2.imread('cxr_ca_train_multi_selected_1' + "/" + filename)
        imgs.append(img)  #存好多圖片,不過因為不是陣列型式所以全部抓出來後要轉型      
       
imgs=np.array(imgs)#轉回array  這是我們要訓練的train 圖片的陣列型式

datagen = ImageDataGenerator(

    rotation_range=5,
   # width_shift_range=0.1,

    fill_mode='nearest')

i = 0
for batch in datagen.flow(imgs, batch_size=1,
                          save_to_dir='C:/Users/Tony/my_final_report/cxr_ca_train_multi_selected_aug_1', save_format='jpg'): 
            i += 1
            if i > 1:
                break  # otherwise the generator would loop indefinitely
```
